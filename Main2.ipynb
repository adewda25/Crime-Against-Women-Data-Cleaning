```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:/PPBI PROJECT/Data/Raw/Cases_under_crime_against_women.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Unnamed: 22'], inplace=True)
```

```python
df.describe()
```

```python
df.drop(columns=[
    "Cases_Compounded_or_Withdrawn",
    "Cases_Withdrawn_by_the_Govt",
    "Cases_withdrawn_by_the_Govt_during_investigation"
], inplace=True)
```

```python
df.rename(columns = {
    'Cases_Acquitted_or_Discharged' : 'Cases_Acquitted',
    'Cases_Pending_Investigation_at_Year_End':'Pending_Investigation_End',
    'Cases_Pending_Trial_at_Year_End':'Pending_Trial_End',
    'Cases_Trials_Completed':'Trials_Completed',
    'Total_For_Trials_Cases' : 'Total_Trial_Cases',
    'Area_Name' : 'State'
},inplace = True)
```

```python
df.drop(columns=[
    'Cases_charge_sheets_were_not_laid_but_Final_Report_submitted',
    'Cases_Declared_False_on_Account_of_Mistake_of_Fact_or_of_Law',
    'Cases_Investigated_Chargesheets+FR_Submitted',
    'Cases_not_Investigated_or_in_which_investigation_was_refused',
    'Cases_Pending_Investigation_from_previous_year',
    'Cases_Pending_Trial_from_the_previous_year',
], inplace=True)
```

```python
df.columns
```

```python
df.duplicated().sum()
```

```python
df.columns=df.columns.str.strip()
```

```python
df.info()
```

```python
numeric_cols = df.select_dtypes(include = 'number').columns
(df[numeric_cols]<0).sum()
```

```python
df = df[(df['Year']>2000) & df['Year']<2025]
```

```python
df.describe()
```

```python
df.columns
```

```python
Top_States = (
    df.groupby('State')['Cases_Reported']
.sum()
.sort_values(ascending = False)
.head(10)
)

print(Top_States)
```

```python
Top_States.plot(kind='bar',figsize=(10,5))
plt.title('Top 10 States with Highest Crimes Against Women')
plt.show()
```

```python
#Crime Trend Over Years
Yearly_Crime = (
    df.groupby('Year')['Cases_Reported']
.sum())

print(Yearly_Crime)
```

```python
Yearly_Crime.plot(kind = 'line',marker ='o')
plt.title('Crime Trend Over Years')
plt.show()
```

```python
Crime_Type = (df.groupby('Sub_Group_Name')['Cases_Reported']
.sum()
.sort_values(ascending = False)
)
print(Crime_Type)
```

```python
 # Most Common Crime Type
Crime_Type.head(10).plot(kind='bar')
plt.title('Top Crime Types')
plt.show()
```

```python
overall_conviction_rate = (
    df['Cases_Convicted'].sum()
    /
    df['Trials_Completed'].sum()
) * 100

print(overall_conviction_rate)
```

```python
state_conviction = (
    df.groupby('State')
      .agg({
          'Cases_Convicted':'sum',
          'Trials_Completed':'sum'
      })
)

state_conviction['Conviction_Rate'] = (
    state_conviction['Cases_Convicted']
    /
    state_conviction['Trials_Completed']
) * 100

state_conviction = state_conviction.sort_values(
    'Conviction_Rate',
    ascending=False
)

print(state_conviction.head(10))
```

```python
pending_trials = (
    df.groupby('State')['Pending_Trial_End']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(pending_trials)
```

```python
chargesheet_rate = (
    df['Cases_Chargesheeted'].sum()
    /
    df['Cases_Reported'].sum()
) * 100

print(chargesheet_rate)
```

```python
Yearly_Crime.plot(kind='line', marker='o', figsize=(10,5))
plt.title('Crime Trend Over Years')
plt.ylabel('Cases Reported')
plt.grid(True)
plt.show()
```

```python
print("Total Cases Reported:", df['Cases_Reported'].sum())
print("Total Cases Acquitted:", df['Cases_Acquitted'].sum())
```
