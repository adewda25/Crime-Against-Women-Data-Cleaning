```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:/PPBI PROJECT/Data/Raw/Arrest_under_crime_against_women.csv")
```

```python
df.columns
```

```python
df.drop(columns=[
    "Persons_in_Custody_or_on_Bail_during_Investigation_at_Year_beginning",
    "Persons_in_Custody_or_on_Bail_during_Investigation_at_Year_end",
    "Persons_in_Custody_or_on_Bail_during_Trial_at_Year_End"
], inplace=True)
```

```python
df.info()
```

```python
rename_map = {
    "Area_Name": "area",
    "Year": "year",
    "Group_Name": "crime_group",
    "Sub_Group_Name": "crime_subgroup",

    "Persons_Acquitted": "acquitted",
    "Persons_against_whom_cases_Compounded_or_Withdrawn": "withdrawn",
    "Persons_Arrested": "arrested",
    "Persons_Chargesheeted": "chargesheeted",
    "Persons_Convicted": "convicted",
    "Persons_Released_or_Freed_by_Police_or_Magistrate_before_Trial_for_want_of_evidence_or_any_other_reason": "released_before_trial",
    "Persons_Trial_Completed": "trial_completed"
}
```

```python
df.rename(columns=rename_map,inplace = True)
```

```python
df.columns.str.strip()
```

```python
df.drop(columns=['Unnamed: 16'], inplace=True)
```

```python
df.rename(columns={
    'crime_group': 'Group_Name',
    'crime_subgroup': 'Sub_Group',
    'area' :'state'
}, inplace=True)
```

```python
df.columns = df.columns.str.title()
```

```python
df.isnull().sum()
```

```python
df.groupby('Year')['Convicted'].sum()
#year wise trend
```

```python
df['High_Crime_Area'] = df['Arrested'] > df['Arrested'].median()
```
