```python
import pandas as pd
```

```python
df = pd.read_csv("D:/PPBI PROJECT/Data/Raw/Specific_purpose_of_kidnapping_and_aduction.csv")
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(columns = [
    "K_A_Female_Upto_10_Years",
    "K_A_Female_10_15_Years",
    "K_A_Female_15_18_Years",
    "K_A_Female_18_30_Years",
    "K_A_Female_30_50_Years",
    "K_A_Female_Above_50_Years",
    "K_A_Male_Upto_10_Years",
    "K_A_Male_10_15_Years",
    "K_A_Male_15_18_Years",
    "K_A_Male_18_30_Years",
    "K_A_Male_30_50_Years",
    "K_A_Male_Above_50_Years"
], inplace=True)
```

```python
df.rename(columns={
    "Area_Name": "State",
    "Sub_Group_Name": "Subgroup_Name",
    "K_A_Cases_Reported": "Cases_Reported",
    "K_A_Female_Total": "Female_Total",
    "K_A_Male_Total": "Male_Total",
    "K_A_Grand_Total": "Total_Cases"
}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
cols = ['Cases_Reported', 'Female_Total', 'Male_Total', 'Total_Cases']

df[cols] = df[cols].fillna(0)
```

```python
df[cols] = df[cols].astype(int)
```

```python
df['Female_Percentage'] = (
    df['Female_Total'] / df['Total_Cases']
) * 100
```

```python
df['Male_Percentage'] = (
    df['Male_Total'] / df ['Total_Cases']
) * 100
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Gender_Gap'] = (df['Female_Total'] - df['Male_Total'])
```

```python
yearly_cases = df.groupby('Year')['Cases_Reported'].sum()

growth = yearly_cases.pct_change() * 100

print(growth)
```

```python
df.drop(columns=[
    'Female_Percentage',
    'Male_Percentage'
], inplace=True)
```

```python
df.to_csv("kidnapping_data.csv", index=False)
```
